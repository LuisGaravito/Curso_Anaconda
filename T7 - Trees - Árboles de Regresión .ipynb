# Árboles de Regresión

```python
import pandas as pd
```

```python
urldata = "https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/boston/Boston.csv"
```

```python
data = pd.read_csv(urldata)
data.head()
```

```python
data.shape
```

```python
colnames = data.columns.values.tolist()
predictors = colnames[:13]
target = colnames[13]
X = data[predictors]
Y = data[target]
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regtree = DecisionTreeRegressor(min_samples_split=30, min_samples_leaf=10, max_depth=5, random_state=0)
```

```python
regtree.fit(X,Y)
```

```python
preds = regtree.predict(data[predictors])
```

```python
data["preds"] = preds
```

```python
data[["preds", "medv"]]
```

```python
from sklearn.tree import export_graphviz
with open("/resources/boston_rtree.dot", "w") as %dotfile:
    export_graphviz(regtree, out_file=dotfile, feature_names=predictors)
    dotfile.close()
    
import os
from graphviz import Source
file =  open("/resources/boston_rtree.dot", "r")
text = file.read()
text
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
cv = KFold(n_splits=X.shape[0], shuffle=True, random_state=1)
scores = cross_val_score(regtree, X, Y, scoring="neg_mean_squared_error", cv = cv, n_jobs=1)
print(scores)
score = np.mean(scores)
print(score)
```

```python
list(zip(predictors,regtree.feature_importances_))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_jobs=2, oob_score=True, n_estimators=10)
forest.fit(X,Y)
```

```python
forest.oob_prediction_
```

```python
data["rforest_pred"] = forest.oob_prediction_
data[["rforest_pred", "medv"]]
```

```python
data["rforest_error2"] = (data["rforest_pred"]-data["medv"])**2
sum(data["rforest_error2"])/len(data)
```

```python
forest.oob_score_
```
