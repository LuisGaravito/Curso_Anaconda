```python
urldata = "https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/iris/iris.csv"
df = pd.read_csv(urldata)
```

```python
X = df.iloc[:,0:4].values
y = df.iloc[:,4].values
X_std = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA as sk_pca
```

```python
acp = sk_pca(n_components=2)
Y = acp.fit_transform(X_std)
```

```python
Y
```

```python
results = []

for name in ('setosa', 'versicolor', 'virginica'):
    result = Scatter(x = Y[y==name,0], y = Y[y==name,1],
                    mode = "markers", name = name,
                    marker = Marker(size=15, line=Line(color="rgba(255,255,255,0.2)", width=0.5),
                                   opacity=0.75))
    results.append(result)
    
data = Data(results)
layout = Layout(xaxis = XAxis(title="CP1", showline=False),
               yaxis = YAxis(title="CP2", showline=False))

fig = Figure(data = Data, layout = layout)
py.iplot(fig)
```
